```python
import os
from dotenv import dotenv_values
from tqdm.notebook import tqdm
import numpy as np

env = dotenv_values()
txtFilemame = "enwiki-20220601-pages-articles-multistream.xml"
txtPath = os.path.join(env['STORE_PATH'],txtFilemame)
n = 4
histOutputPath = os.path.join(env['STORE_PATH'],"hist89-{}gram.mm".format(n))
print(f"{txtPath=}")
print(f"{histOutputPath=}")
```

```python
txt = np.memmap(txtPath,mode='r',dtype=np.uint8)
from numpy.lib.stride_tricks import as_strided
```

```python
import pickle
from collections import defaultdict
hist = defaultdict(int)
import numpy as np
from scipy.sparse import dok_array
```

```python
%%time
# hist = dok_array((0xffffffff, 1), dtype=np.uint64)
sliding_window_view = as_strided(txt, shape=(len(txt)//3,4), strides=(1,1)).view(np.uint32)
blocksize = 0x10000000 #items
print("iterations: ",int(len(sliding_window_view)/blocksize)+1)
for i, in tqdm(np.ndindex((int(len(sliding_window_view)/blocksize)+1,))):
    slice = lambda:(sliding_window_view[i*blocksize:(i+1)*blocksize]&0x00ffffff).reshape(-1)
    # for ngram in slice:
        # hist[ngram,0]+=1
    %time index,counts = np.unique(slice(), return_counts=True)
    for i in range(len(index)):
        hist[index[i]]+=counts[i]
with open('saved_dictionary.pkl', 'wb') as f:
    pickle.dump(hist, f)
```
